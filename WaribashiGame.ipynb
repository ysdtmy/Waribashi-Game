## Import Library

```python
from waribashi import *
import pickle
```

## Import Qtable

```python
with open('./Qtables/Qtable_iter1000000.bin', 'rb') as f:
    Qtable = pickle.load(f)
```

```python
Qtable
```

## Initialize  Environment

```python
e = WaribashiGame()
```

## Initialize agents

### Greedy Q-learning agent

```python
p0 = Q_agent(epsilon=0)
```

### Random agent

```python
p1 = Q_agent(epsilon=1)
```

### Playable agent

```python
p2 = human()
```

## Greedy Q-leaning VS Random

```python
GM = GameMaster(e, p0, p1,Qtable=Qtable)
result = GM.play(iter_num=1000, silent=False)
```

### Result

```python
print('draw:' +str(result[-1]))
print('player0 wins :' +str(result[0]))
print('player1 wins:' +str(result[1]))
```
